```python
grouped = eda.groupby(['Port_IDs', 'symbol'])

profits = []

for (port_id, symbol), trades in grouped:
    # Separate BUY and SELL trades
    buy_trades = trades[(trades['side'] == 'BUY') & (trades['activeBuy'] == True)]
    sell_trades = trades[(trades['side'] == 'SELL')]

    # Sort by time
    buy_trades = buy_trades.sort_values(by='time')
    sell_trades = sell_trades.sort_values(by='time')

    # Match BUY and SELL trades
    for _, buy in buy_trades.iterrows():
        for _, sell in sell_trades.iterrows():
            if sell['qty'] >= buy['qty']:
                # Calculate realized profit
                profit = (sell['price'] - buy['price']) * buy['qty']
                net_profit = profit - buy['fee'] - sell['fee']

                # Append to results
                profits.append({
                    'Port_IDs': port_id,
                    'symbol': symbol,
                    'buy_time': buy['time'],
                    'sell_time': sell['time'],
                    'buy_price': buy['price'],
                    'sell_price': sell['price'],
                    'qty': buy['qty'],
                    'realizedProfit': net_profit
                })

                # Reduce the qty in SELL trade
                sell['qty'] -= buy['qty']
                break

# Convert profits to DataFrame
profit_df = pd.DataFrame(profits)

# Display results
print(profit_df)
```

```python
# Calculate profit per trade
profit_analysis = df.copy()
profit_analysis['trade_value'] = profit_analysis['price'] * profit_analysis['qty']

# Group by Port_IDs, symbol, and side to calculate buy/sell values
portfolio_profit = profit_analysis.groupby(['Port_IDs', 'symbol', 'side']).agg({
    'trade_value': 'sum',
    'fee': 'sum',
    'qty': 'sum'
}).reset_index()

# Pivot the data to get BUY and SELL columns
portfolio_summary = portfolio_profit.pivot(
    index=['Port_IDs', 'symbol'],
    columns='side',
    values=['trade_value', 'fee', 'qty']
).fillna(0)

# Calculate total profit
portfolio_summary['total_profit'] = (
    portfolio_summary[('trade_value', 'SELL')] - 
    portfolio_summary[('trade_value', 'BUY')] - 
    portfolio_summary[('fee', 'BUY')] - 
    portfolio_summary[('fee', 'SELL')]
)

# Reset index and sort by profit
portfolio_summary = portfolio_summary.reset_index().sort_values('total_profit', ascending=False)

# Display overall statistics
print("\nProfit Statistics by Port_ID and Symbol:")
print(portfolio_summary['total_profit'].describe())

# Optional: Display top 10 most profitable combinations
print("\nTop 10 Most Profitable Port_ID-Symbol Combinations:")
print(portfolio_summary[['Port_IDs', 'symbol', 'total_profit']].head(10))
```

```python
# Create a DataFrame from port_profit series
port_profit_df = pd.DataFrame(port_profit).reset_index()
port_profit_df.columns = ['Port_IDs', 'total_profit']

# Calculate total investment using the original portfolio_summary DataFrame
buy_trades = profit_analysis[profit_analysis['side'] == 'BUY'].groupby('Port_IDs').agg({
    'trade_value': 'sum'
}).reset_index()
buy_trades.columns = ['Port_IDs', 'total_investment']

# Merge investment data with profit data
port_profit_df = port_profit_df.merge(buy_trades, on='Port_IDs')

# Calculate ROI
port_profit_df['ROI'] = (port_profit_df['total_profit'] / port_profit_df['total_investment']) * 100

# Add summary statistics
port_profit_df['profit_rank'] = port_profit_df['total_profit'].rank(ascending=False)
port_profit_df['profit_percentile'] = port_profit_df['total_profit'].rank(pct=True)

# Sort by ROI descending
port_profit_df = port_profit_df.sort_values('ROI', ascending=False)

# Display the first few rows with formatted columns
print("\nPortfolio Performance Analysis:")
print(port_profit_df[['Port_IDs', 'total_investment', 'total_profit', 'ROI', 'profit_rank', 'profit_percentile']].head())
```
